```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np

def load_image(path_image):
    img = cv2.imread("/content/20221013_093252.jpg")
    return img[...,::-1] 

img =  cv2.imread("/content/20221013_093252.jpg")

#Rojo
redBajo1 = np.array([0, 100, 20], np.uint8)
redAlto1 = np.array([7, 255, 255], np.uint8)
redBajo2=np.array([160, 100, 20], np.uint8)
redAlto2=np.array([180, 255, 255], np.uint8)
#Naranja
orangeBajo1 = np.array([7.5, 100, 20], np.uint8)
orangeAlto1 = np.array([20, 255, 255], np.uint8)

#amarillo
yellowBajo1 = np.array([20, 100, 20], np.uint8)
yellowAlto1 = np.array([36, 255, 255], np.uint8)

#Blanco
whiteBajo1 = np.array([0, 0, 100], np.uint8)
whiteAlto1 = np.array([255, 20, 255], np.uint8)


frameHSV = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

maskRed1 = cv2.inRange(frameHSV, redBajo1, redAlto1)
maskRed2 = cv2.inRange(frameHSV, redBajo2, redAlto2)
maskRed = cv2.add(maskRed1, maskRed2)
maskRedvis = cv2.bitwise_and(load_image(img),load_image(img),mask=maskRed)

maskorange = cv2.inRange(frameHSV, orangeBajo1,orangeAlto1)
maskorangevis = cv2.bitwise_and(load_image(img),load_image(img),mask=maskorange)

maskyellow = cv2.inRange(frameHSV, yellowBajo1,yellowAlto1)
maskyellowvis = cv2.bitwise_and(load_image(img),load_image(img),mask=maskyellow)

maskwhite = cv2.inRange(frameHSV,whiteBajo1,whiteAlto1)
maskwhitevis = cv2.bitwise_and(load_image(img),load_image(img),mask=maskwhite)

fig, axs = plt.subplots(1,5,figsize=(20,5))
axs[0].imshow(load_image(img))
axs[1].imshow(maskRedvis )
axs[1].set_title('Rojo')
axs[2].imshow(maskorangevis)
axs[2].set_title('Naranja')
axs[3].imshow(maskyellowvis)
axs[3].set_title('Amarillo')
axs[4].imshow(maskwhitevis)
axs[4].set_title('Blanco')


```
